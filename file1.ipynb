```python
import joblib
import pandas as pd

# =============================
# LOAD MODEL
# =============================
model = joblib.load("car_claim_numeric_model.pkl")

# =============================
# LOAD DATA TEMPLATE
# =============================
train_df = pd.read_csv("train.csv")

# Create base input structure
input_data = train_df.drop(columns=["is_claim"]).iloc[0:1].copy()

# =============================
# MODIFY CUSTOMER VALUES
# =============================
input_data["age_of_policyholder"] = 35
input_data["age_of_car"] = 5

# =============================
# PREDICT CLAIM PROBABILITY
# =============================
claim_prob = model.predict_proba(input_data)[:, 1][0]

# =============================
# DECISION LOGIC
# =============================
if claim_prob < 0.30:
    risk_level = "Low Risk"
    approval_likelihood = "High"
    decision = "Auto Approve"
elif claim_prob < 0.60:
    risk_level = "Medium Risk"
    approval_likelihood = "Moderate"
    decision = "Manual Review"
else:
    risk_level = "High Risk"
    approval_likelihood = "Low"
    decision = "Investigate"

# =============================
# OUTPUT
# =============================
print("\n===== CLAIM ASSESSMENT RESULT =====")
print(f"Claim Probability : {claim_prob*100:.2f}%")
print(f"Risk Level        : {risk_level}")
print(f"Approval Chance   : {approval_likelihood}")
print(f"Decision          : {decision}")
```
